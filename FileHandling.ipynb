### Task 1 : Write Sales Record To A File

```python
sales = [1200, 450, 980, 1500, 3000]

# Write sales data to file
with open("sales_data.txt", "w") as file:
    for amount in sales:
        file.write(str(amount) + "\n")

# Reopen and print contents
print("Task 1 Output:")
with open("sales_data.txt", "r") as file:
    print(file.read())
```

### Task 2: Read File in Different Ways

```python
print("\nTask 2 Output:")

# 1. Read entire file
with open("sales_data.txt", "r") as file:
    content = file.read()
    print("Using read():")
    print(content)

# 2. Read first line
with open("sales_data.txt", "r") as file:
    first_line = file.readline().strip()
    print("Using readline():")
    print(first_line)

# 3. Read all lines and convert to integers
with open("sales_data.txt", "r") as file:
    sales_list = [int(line.strip()) for line in file.readlines()]

print("Using readlines() converted to integers:")
print(sales_list)
```

### Task 3: Append New Sales

```python
new_sales = [5000, 2500, 1700]

# Append new sales data
with open("sales_data.txt", "a") as file:
    for amount in new_sales:
        file.write(str(amount) + "\n")

# Reopen and print updated file
print("\nTask 3 Output:")
with open("sales_data.txt", "r") as file:
    updated_content = file.read()
    print(updated_content)

# Optional: Count total lines
with open("sales_data.txt", "r") as file:
    total_lines = len(file.readlines())

print("Total number of lines:", total_lines)
```

### Task 4: Generate Summary Report from File

```python
# Read sales data from file
with open("sales_data.txt", "r") as file:
    sales = [int(line.strip()) for line in file.readlines()]

# Calculate summary
total_sales = sum(sales)
highest_sale = max(sales)
lowest_sale = min(sales)
average_sale = total_sales / len(sales)

# Print report
print("SALES SUMMARY REPORT")
print("Total Sales :", total_sales)
print("Highest Sale:", highest_sale)
print("Lowest Sale :", lowest_sale)
print("Average Sale:", round(average_sale, 2))
```

### Task 5: Create Product Info File

```python
# Ask user for number of products
n = int(input("How many products do you want to enter? "))

# Write product details to file
with open("products.txt", "w") as file:
    for i in range(n):
        name = input("Enter product name: ")
        price = input("Enter product price: ")
        file.write(f"{name} | {price}\n")

# Read and display file contents
print("\n===== PRODUCT LIST =====")
with open("products.txt", "r") as file:
    for line in file:
        product, price = line.strip().split("|")
        print(f"Product: {product.strip()}  |  Price: ₹{price.strip()}")
```

### Task 6: Read File Safely

```python
import os

filename = input("Enter filename: ")

if os.path.exists(filename):
    with open(filename, "r") as file:
        print("\nFile Contents:")
        print(file.read())
else:
    print("File not found. Please check the filename.")
```
