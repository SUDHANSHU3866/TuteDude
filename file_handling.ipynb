## File Handling

### Read the content from My text file in the system.

```python
# open() --> opening your file.
# First Parameter --> File_name
# Second Parameter --> Mode
# Mode --> It decides the way in which we are working with this file.
# Mode --> r --> Reading the content from the file.
# Mode --> w --> Write the content into the file.
# Mode --> r+ --> Read the content first and the write the content into the file.
# Mode --> w+ --> Writing the content first and then read from it
```

```python
# Reading Your File :
# read() --> Read the content from the file.
```

```python
f = open("example.txt","r")
content = f.read()
print(content)
f.close()
```

### Writing the content into the file.

### write() --> it overwrites the content into the text file.
### Mode --> w --> Content will be overwritten.
### Mode --> a --> Content will be added from the end.

```python
# write() --> helps us out in writing the content into the file
```

```python
f = open('example.txt','w')
f.write('Hello We are not writing the content into our File.')
f.close()
```

```python
# Appending the content into our file

f = open('example.txt','a')
f.write('\nWe are learning about GenAi')
f.close()
```

### with --> It will be used for first of all opening the file as a object and there is no requirement for closing the file as well.

```python
with open('example.txt','r') as f:
    content_data = f.read()
    print(content_data)
```

```python
with open('example.txt','w') as file:
    file.write("We are exploring about with command in python File Handling.")
```

```python
with open('example.txt','a') as file:
    file.write("\nHello we are exploring appendd mode as well.")
```

```python
# Wite the content into our file line by line --> writelines()
```

```python
list = ["First Line\n","Second Line\n","Third Line\n"]
with open('example.txt','w') as data:
    data.writelines(list)
```

```python
# Reading the total lines , total words , total character from a file ---> readlines()

def count_lines(file):
    with open(file,'r') as files:
        lines = files.readlines()
        total_lines = len(lines)
        total_words = sum(len(line.split()) for line in lines)
        total_characters = sum(len(line) for line in lines)
        
    return total_lines,total_words,total_characters

file_path = 'total_data.txt'
Total_lines,Total_words,Total_characters = count_lines(file_path)
print(f"Total Lines are : {Total_lines} and total words are : {Total_words} and total characters are : {Total_characters}")
```

```python
# Writing the content into a file and then after writing the content into that file we are 
# going to read the content from that file as well. --> seek() and w+
```

#### w+ ---> First of all as w is their so it states that first writing operation will be done and then read operation will be done.

#### seek() ---> It is going to help us out with reading the content after writing from the start only.  

```python
with open('total_data.txt','w+') as write_read_file:
    write_read_file.write("\nWe are looking into w+ operation.")
    write_read_file.write("\nIt means writing and then reading.")
    write_read_file.seek(0)
    content_complete_data = write_read_file.read()
    print(content_complete_data)
```
